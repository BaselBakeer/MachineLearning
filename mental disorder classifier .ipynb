```python
import pandas as pd 
```

```python
df=pd.read_excel('Mental disorder symptoms.xlsx')
```

```python
df
```

```python
df.isnull()
```

```python
df.info()
```

```python
df1=pd.DataFrame(pd.get_dummies(df['Disorder']))
```

```python
df1
```

```python
df=pd.concat([df1,df],axis=1)
```

```python
df
```

```python
df.drop('Disorder',axis=1,inplace=True)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(['ADHD','ASD','Loneliness','MDD','OCD','PDD','PTSD','bipolar','eating disorder','psychotic deprission','anexiety','sleeping disorder'],axis=1)
```

```python
y=df.filter(['ADHD','ASD','Loneliness','MDD','OCD','PDD','PTSD','bipolar','eating disorder','psychotic deprission','anexiety','sleeping disorder'],axis=1)
```

```python
y
```

```python
x
```

```python
x_train , x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators=100)
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
model.score(x_test,y_test)
```
