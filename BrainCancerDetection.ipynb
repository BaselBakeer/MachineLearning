```python
import pandas as pd 
```

```python
df=pd.read_excel('BrainCancerData.xlsx')
```

```python
df
```

```python
df['class label']=pd.cut(df['class label'],5, labels=["1st", "2nd", "3rd","4th","5th"])
```

```python
df
```

```python
df1=pd.DataFrame(pd.get_dummies(df['class label']))
df = pd.concat([df1, df], axis=1)
```

```python
df
```

```python
df.drop('class label',axis=1,inplace=True )
```

```python
df
```

```python
df1=df.iloc[0:200]
```

```python
df1
```

```python
x=df1.drop(['1st','2nd','3rd','4th','5th'],axis=1)
y=df1.filter(['1st','2nd','3rd','4th','5th'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier(n_estimators=200)
model1.fit(x_train, y_train)
```

```python
model1.score(x_test,y_test)
```
